# safepython cli

```
#| export
from fastcore.script import call_parse
from pathlib import Path
from safepyrun.core import RunPython

import asyncio
```

```
#| export
@call_parse
def main(path: str):
    "Run a python script file in the safepyrun sandbox"
    pyrun = RunPython()
    asyncio.run(pyrun(Path(path).read_text()))
```
